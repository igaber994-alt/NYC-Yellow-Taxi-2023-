# The NYC Yellow Taxi System

Before starting the technical part of the project, it is important to understand what the NYC Yellow Taxi system is and how the trip data is created.

## What is NYC Yellow Taxi?

NYC Yellow Taxi is one of the official taxi services operating in New York City.

These taxis are easy to recognize because they are yellow and licensed to operate in the city. They provide transportation for passengers from one location to another, similar to normal taxis, but the system is more organized and highly regulated.

Each taxi trip creates useful operational data that can later be analyzed to understand demand, revenue, trip behavior, and transportation efficiency.

---

## How is a Taxi Trip Recorded?

Most trip information is recorded automatically by the taxi systems inside the vehicle.

The driver does not manually write every detail of the trip. Instead, the taxi has systems such as:

- **Meter system**
- **Payment system**
- **Location zone system**

These systems help record important information about each trip.

---

## Meter System

The **meter system** is the taxi meter that calculates and records the trip.

When the trip starts, the meter records the pickup time.  
When the trip ends, the meter records the dropoff time.

The meter system can also record or calculate:

- Trip start time
- Trip end time
- Trip distance
- Base fare
- Extra charges
- Taxes and surcharges

For example, if a passenger starts a trip at 8:00 PM and ends the trip at 8:20 PM, the system records these two timestamps and uses them to calculate trip duration.

---

## Payment System

The **payment system** records how the passenger paid for the trip.

The payment type may be:

- Credit card
- Cash
- No charge
- Dispute
- Unknown
- Voided trip

If the passenger pays by credit card and adds a tip, the tip amount is recorded in the dataset.

However, if the passenger gives a cash tip, it may not appear in the data. This is important when analyzing tipping behavior.

---

## Pickup and Dropoff Locations

The dataset does not usually store the exact street address of the passenger.

Instead, New York City is divided into taxi zones. Each zone has a unique ID.

The dataset uses:

- **PULocationID** = Pickup Location ID
- **DOLocationID** = Dropoff Location ID

For example:

# Step 1 — Data Ingestion into Bronze Layer

## What We Are Doing

In this step, we ingest the full 2023 NYC Yellow Taxi dataset from the raw data volume into a Spark DataFrame.

The process includes:

- Starting a Spark session
- Reading all monthly Parquet files from the raw data volume
- Loading 12 months of taxi trip data
- Standardizing selected column data types
- Adding metadata columns
- Combining all monthly files into one Bronze DataFrame
- Previewing the loaded dataset

````python
from pyspark.sql import SparkSession 
from pyspark.sql.functions import *
from pyspark.sql.types import *

spark = SparkSession.builder.appName("final project").getOrCreate()
````

````python
from pyspark.sql.functions import current_timestamp, col, lit
from functools import reduce

raw_path = "/Volumes/main/default/raw_data/" 

files = dbutils.fs.ls(raw_path) # بتعرض الملفات الموجودة في المسار المحدد.
# parquet_files = [f.path for f in files if f.name.endswith(".parquet")]
# List Comprehension.
# For each file in files, if the file name ends with .parquet, take its path.
parquet_files = []

for f in files:
    if f.name.endswith(".parquet"):
        parquet_files.append(f.path)

dfs = []

for file_path in parquet_files:
    print(f"Reading: {file_path}")
    
    temp_df = spark.read.parquet(file_path)
    


    temp_df = (
        temp_df
        .withColumn("passenger_count", col("passenger_count").cast("double"))
        .withColumn("RatecodeID", col("RatecodeID").cast("double"))
        .withColumn("airport_fee", col("airport_fee").cast("double"))
        .withColumn("_source_file", lit(file_path))
    )
    # ده مهم جدًا في حاجة اسمها Data Lineage، يعني تتبع مصدر البيانات.
    dfs.append(temp_df)

# تجمع كل DataFrames مع بعض واحدة واحدة.reduce
df_bronze = reduce(
    lambda df1, df2: df1.unionByName(df2, allowMissingColumns=True),
    dfs
)
# لو عمود موجود في df ومش موجود في التاني 
# → كمّله بـ NULL

print(f"Total rows loaded: {df_bronze.count():,}")
display(df_bronze.limit(10))
````

## Detailed Percentile Profiling

````python
from pyspark.sql import functions as F

df_bronze_profile = (
    df_bronze_delta
    .withColumn(
        "trip_duration_minutes",
        (
            F.unix_timestamp("tpep_dropoff_datetime") -
            F.unix_timestamp("tpep_pickup_datetime")
        ) / 60
    )
    .withColumn(
        "average_speed_mph",
        F.when(
            F.col("trip_duration_minutes") > 0,
            F.col("trip_distance") / (F.col("trip_duration_minutes") / 60)
        ).otherwise(None)
    )
)

def create_percentile_profile(df, columns, accuracy=10000):
    percentile_array = "array(0.01, 0.25, 0.50, 0.75, 0.95, 0.99, 0.999)"

    agg_exprs = [
        F.expr(
            f"percentile_approx(`{c}`, {percentile_array}, {accuracy})"
        ).alias(c)
        for c in columns
    ]

    wide_df = df.agg(*agg_exprs)

    stack_expr = ", ".join([f"'{c}', `{c}`" for c in columns])

    long_df = wide_df.selectExpr(
        f"stack({len(columns)}, {stack_expr}) as (metric, percentiles)"
    )

    result_df = long_df.select(
        "metric",
        F.element_at("percentiles", 1).alias("p01"),
        F.element_at("percentiles", 2).alias("p25"),
        F.element_at("percentiles", 3).alias("p50_median"),
        F.element_at("percentiles", 4).alias("p75"),
        F.element_at("percentiles", 5).alias("p95"),
        F.element_at("percentiles", 6).alias("p99"),
        F.element_at("percentiles", 7).alias("p999")
    )

    return result_df
````

````python
financial_columns = [
    "fare_amount",
    "total_amount",
    "extra",
    "mta_tax",
    "tip_amount",
    "tolls_amount",
    "improvement_surcharge",
    "congestion_surcharge",
    "airport_fee"
]

financial_percentile_profile = create_percentile_profile(
    df_bronze_profile,
    financial_columns,
    accuracy=10000
)

display(financial_percentile_profile)
````

````python
trip_metric_columns = [
    "trip_distance",
    "trip_duration_minutes",
    "average_speed_mph"
]

trip_metric_percentile_profile = create_percentile_profile(
    df_bronze_profile,
    trip_metric_columns,
    accuracy=10000
)

display(trip_metric_percentile_profile)
````

## Financial Problems


````python
from pyspark.sql import functions as F

checks = [
    ("invalid_fare_amount", (F.col("fare_amount") <= 0) | (F.col("fare_amount") > 500)),
    ("invalid_total_amount", (F.col("total_amount") <= 0) | (F.col("total_amount") > 600)),
    ("negative_tip_amount", F.col("tip_amount") < 0),

    ("invalid_extra", (F.col("extra") < 0) | (F.col("extra") > 20)),
    ("invalid_mta_tax", (F.col("mta_tax") < 0) | (F.col("mta_tax") > 1)),
    ("invalid_tolls_amount", (F.col("tolls_amount") < 0) | (F.col("tolls_amount") > 20)), # 1 8%
    ("negative_improvement_surcharge", F.col("improvement_surcharge") < 0),
    ("negative_congestion_surcharge", F.col("congestion_surcharge") < 0),
    ("negative_airport_fee", F.col("airport_fee") < 0)
]

money_quality_counts = df_bronze_delta.select(
    *[
        F.sum(condition.cast("int")).alias(check_name)
        for check_name, condition in checks
    ]
)

money_quality_profile = money_quality_counts.select(
    F.explode(
        F.array(
            *[
                F.struct(
                    F.lit(check_name).alias("quality_check"),
                    F.col(check_name).alias("records"),
                    F.round((F.col(check_name) / total_rows * 100), 6).alias("percentage")
                )
                for check_name, condition in checks
            ]
        )
    ).alias("result")
).select(
    "result.quality_check",
    "result.records",
    "result.percentage"
)

display(money_quality_profile)
````

| Issue | Summary | Silver Layer Action |
|---|---|---|
| Financial data problems | The financial columns contain invalid, negative, and extreme values, including invalid fares, invalid total amounts, negative tips, and abnormal taxes, tolls, fees, and surcharges. These may represent refunds, disputes, corrections, voided trips, or abnormal transactions. | Keep normal completed trips using financial rules: `0 < fare_amount <= 500`, `0 < total_amount <= 600`, `tip_amount >= 0`, `0 <= extra <= 20`, `0 <= mta_tax <= 1`, `0 <= tolls_amount <= 100`, `improvement_surcharge >= 0`, `congestion_surcharge >= 0`, and `airport_fee >= 0`. |

### Key Insight

I focused first on `fare_amount` and `total_amount` because they are the strongest financial indicators in the project. Then I added stricter component-level checks for fees, taxes, tolls, and surcharges to improve the financial quality of the Silver layer.

These additional component filters removed only a very small number of records, so they improved data quality without materially changing the dataset size.

````python
df_speed_profile = df_duration_profile.withColumn(
    "average_speed_mph",
    when(
        col("trip_duration_minutes") > 0,
        col("trip_distance") / (col("trip_duration_minutes") / 60)
    ).otherwise(None)
)

display(
    df_speed_profile
    .select("average_speed_mph")
    .summary()
)

speed_quality_profile = df_speed_profile.select(
    sum(
        (
            (col("average_speed_mph") <= 0) |
            (col("average_speed_mph") > 100)
        ).cast("int")
    ).alias("invalid_speed_records"),

    (
        sum(
            (
                (col("average_speed_mph") <= 0) |
                (col("average_speed_mph") > 100)
            ).cast("int")
        ) / total_rows * 100
    ).alias("invalid_speed_percentage")
)

display(speed_quality_profile)
````


| Issue | Problem | Silver Layer Action |
|---|---|---|
| Unrealistic average speed | Most trips have reasonable city speeds, with median speed around `9.58 mph`. However, some records show impossible speeds, with maximum speed above `2.5 million mph`. This may indicate errors in distance, duration, or timestamps. | Keep only realistic speeds: `0 < average_speed_mph <= 100` |

## Financial Consistency Check

````python
from pyspark.sql.functions import col, sum, abs, round, coalesce, lit

tolerance = 0.01

df_financial_consistency = df_bronze_delta.withColumn(
    "calculated_total_amount",
    coalesce(col("fare_amount"), lit(0)) +
    coalesce(col("extra"), lit(0)) +
    coalesce(col("mta_tax"), lit(0)) +
    coalesce(col("tip_amount"), lit(0)) +
    coalesce(col("tolls_amount"), lit(0)) +
    coalesce(col("improvement_surcharge"), lit(0)) +
    coalesce(col("congestion_surcharge"), lit(0)) +
    coalesce(col("airport_fee"), lit(0))
).withColumn(
    "total_amount_difference",
    round(col("total_amount") - col("calculated_total_amount"), 4)
).withColumn(
    "is_total_amount_inconsistent",
    abs(col("total_amount_difference")) > tolerance
)

financial_consistency_profile = df_financial_consistency.select(
    sum(col("is_total_amount_inconsistent").cast("int")).alias("inconsistent_total_amount_records"),
    (
        sum(col("is_total_amount_inconsistent").cast("int")) / total_rows * 100
    ).alias("inconsistent_total_amount_percentage")
)

display(financial_consistency_profile)


display(
    df_financial_consistency
    .filter(col("is_total_amount_inconsistent") == True)
    .select(
        "fare_amount",
        "extra",
        "mta_tax",
        "tip_amount",
        "tolls_amount",
        "improvement_surcharge",
        "congestion_surcharge",
        "airport_fee",
        "calculated_total_amount",
        "total_amount",
        "total_amount_difference",
        "_source_file"
    )
    .limit(20)
)
````


| Issue | Result | Action |
|---|---:|---|
| Inconsistent total amount | `10,227,219` records, about `26.70%` | Do not delete directly. Create a flag: `is_total_amount_inconsistent` |

### Key Insight

Many differences appear as repeated values like `-1.75`, `-2.50`, and `-4.25`, which are likely related to fee components such as `airport_fee` and `congestion_surcharge`.

So, these records are not necessarily corrupted. Flagging them is better than deleting a large part of the dataset.


## Financial Consistency Flow

| Step | Purpose | What We Do |
|---|---|---|
| Profiling | Discover and understand the problem | Check how many records are financially inconsistent and study the difference pattern |
| Silver Layer | Store data quality information | Keep the records, but create a flag called `is_total_amount_inconsistent` |

### Why This Is Logical

In Profiling, we only investigate the issue.  
In Silver, we save the quality flag without deleting many records.  
In Feature Engineering, we turn the technical flag into a business-friendly label for dashboards and reports.

### Key Insight

## Silver Preparation 

This step prepares the Bronze data before applying the final Silver filters.

It does not only clean obvious issues like duplicates and missing optional fees, but also creates important helper columns such as:

- `pickup_year`
- `pickup_month`
- `pickup_date`
- `trip_duration_minutes`
- `average_speed_mph`
- `is_total_amount_inconsistent`

### Key Insight

The purpose of this step is to make the dataset ready for accurate filtering.  
Instead of applying cleaning rules directly on raw columns, we first create clear helper columns that support time filtering, duration checks, speed validation, and financial consistency tracking.

This makes the Silver cleaning process more logical, explainable, and reliable.

````python
tolerance = 0.01

df_silver_prepared = (
    df_bronze_delta

    # Remove exact duplicate rows first
    .dropDuplicates()

    # Standardize nullable fee fields
    .withColumn(
        "airport_fee",
        F.coalesce(F.col("airport_fee").cast("double"), F.lit(0.0))
    )
    .withColumn(
        "congestion_surcharge",
        F.coalesce(F.col("congestion_surcharge").cast("double"), F.lit(0.0))
    )

    # Handle optional categorical nulls
    .withColumn(
        "store_and_fwd_flag",
        F.coalesce(F.col("store_and_fwd_flag"), F.lit("Unknown"))
    )

    # Time helper columns needed for filtering and partitioning
    .withColumn("pickup_year", F.year("tpep_pickup_datetime"))
    .withColumn("pickup_month", F.month("tpep_pickup_datetime"))
    .withColumn("pickup_date", F.to_date("tpep_pickup_datetime"))

    # Trip duration in minutes
    .withColumn(
        "trip_duration_minutes",
        (
            F.unix_timestamp("tpep_dropoff_datetime")
            - F.unix_timestamp("tpep_pickup_datetime")
        ) / 60
    )

    # Average speed in miles per hour
    .withColumn(
        "average_speed_mph",
        F.when(
            F.col("trip_duration_minutes") > 0,
            F.col("trip_distance") / (F.col("trip_duration_minutes") / 60)
        ).otherwise(None)
    )

    # Financial consistency check
    .withColumn(
        "calculated_total_amount",
        F.coalesce(F.col("fare_amount"), F.lit(0.0)) +
        F.coalesce(F.col("extra"), F.lit(0.0)) +
        F.coalesce(F.col("mta_tax"), F.lit(0.0)) +
        F.coalesce(F.col("tip_amount"), F.lit(0.0)) +
        F.coalesce(F.col("tolls_amount"), F.lit(0.0)) +
        F.coalesce(F.col("improvement_surcharge"), F.lit(0.0)) +
        F.coalesce(F.col("congestion_surcharge"), F.lit(0.0)) +
        F.coalesce(F.col("airport_fee"), F.lit(0.0))
    )
    .withColumn(
        "total_amount_difference",
        F.round(F.col("total_amount") - F.col("calculated_total_amount"), 4)
    )
    .withColumn(
        "is_total_amount_inconsistent",
        F.abs(F.col("total_amount_difference")) > tolerance
    )
)
````

## Silver Cleaning 

This step creates the final clean Silver dataset by applying all data quality rules from the Bronze profiling results.

The filters remove records with:

- Missing pickup or dropoff timestamps
- Dates outside 2023
- Invalid trip time logic
- Zero, negative, or extreme trip distance
- Invalid fare and total amount values
- Negative or unrealistic financial components
- Negative tips
- Invalid passenger count
- Invalid RatecodeID
- Invalid payment types
- Invalid trip duration
- Unrealistic average speed

### Key Insight

The Silver Layer keeps only valid, realistic, and completed taxi trips.

These rules make the dataset reliable for analysis, dashboards, and machine learning because the cleaning decisions are based on profiling results, not random assumptions.

````python
df_silver = (
    df_silver_prepared

    # Required timestamps
    .filter(F.col("tpep_pickup_datetime").isNotNull())
    .filter(F.col("tpep_dropoff_datetime").isNotNull())

    # Keep only 2023 trips
    .filter(F.col("pickup_year") == 2023)

    # Valid trip time
    .filter(F.col("tpep_dropoff_datetime") > F.col("tpep_pickup_datetime"))

    # Valid distance
    .filter((F.col("trip_distance") > 0) & (F.col("trip_distance") <= 100))

    # Valid fare and passenger charge
    .filter((F.col("fare_amount") > 0) & (F.col("fare_amount") <= 500))
    .filter((F.col("total_amount") > 0) & (F.col("total_amount") <= 600))

    # Valid financial components
    .filter((F.col("extra") >= 0) & (F.col("extra") <= 20))
    .filter((F.col("mta_tax") >= 0) & (F.col("mta_tax") <= 1))
    .filter((F.col("tolls_amount") >= 0) & (F.col("tolls_amount") <= 100))
    .filter(F.col("improvement_surcharge") >= 0)
    .filter(F.col("congestion_surcharge") >= 0)
    .filter(F.col("airport_fee") >= 0)

    # Valid tip
    .filter((F.col("tip_amount") >= 0) & (F.col("tip_amount") <= 100))

    # Valid passenger count
    .filter((F.col("passenger_count") >= 1) & (F.col("passenger_count") <= 6))

    # Valid RatecodeID
    .filter((F.col("RatecodeID") >= 1) & (F.col("RatecodeID") <= 6))

    # Valid paid trip payment types
    .filter(F.col("payment_type").isin(1, 2))

    # Valid duration
    .filter((F.col("trip_duration_minutes") > 0) & (F.col("trip_duration_minutes") <= 240))

    # Valid speed
    .filter((F.col("average_speed_mph") > 0) & (F.col("average_speed_mph") <= 100))

    # Cast selected categorical columns after filtering
    .withColumn("VendorID", F.col("VendorID").cast("int"))
    .withColumn("passenger_count", F.col("passenger_count").cast("int"))
    .withColumn("RatecodeID", F.col("RatecodeID").cast("int"))
    .withColumn("PULocationID", F.col("PULocationID").cast("int"))
    .withColumn("DOLocationID", F.col("DOLocationID").cast("int"))
    .withColumn("payment_type", F.col("payment_type").cast("int"))
)
````

## Feature Engineering Step

In this step, we create analysis and machine learning features from the cleaned Silver dataset.

| Feature Group | Examples | Purpose |
|---|---|---|
| Time features | `pickup_hour`, `month_name`, `day_type`, `time_period`, `is_rush_hour` | Analyze demand by time, day, month, and rush hours |
| Distance features | `distance_bucket`, `fare_per_mile`, `total_per_mile` | Understand trip length and cost per mile |
| Duration features | `duration_bucket` | Classify trips based on duration |
| Speed features | `speed_category` | Analyze trip speed and traffic behavior |
| Tip features | `tip_base_amount`, `tip_percentage`, `tip_category` | Analyze tipping behavior, mainly for credit card trips |
| Labels | `payment_type_label`, `ratecode_label` | Convert numeric codes into readable business labels |
| Airport feature | `is_airport_trip` | Identify airport-related trips |
| Route feature | `route_pair` | Analyze pickup-to-dropoff route patterns |
| Financial label | `financial_consistency_label` | Convert financial consistency flag into a readable label |

### Key Insight

Feature engineering transforms the cleaned Silver data into a business-ready dataset for Gold analysis, dashboards, and machine learning.

````python
df_features = (
    df_silver_delta

    # =========================
    # Time-based features
    # =========================
    .withColumn(
        "pickup_hour",
        F.hour(F.col("tpep_pickup_datetime"))
    )

    .withColumn(
        "pickup_day_name",
        F.date_format(F.col("tpep_pickup_datetime"), "EEEE")
    )

    .withColumn(
        "pickup_day_of_week",
        F.dayofweek(F.col("tpep_pickup_datetime"))
    )

    .withColumn(
        "month_name",
        F.date_format(F.col("pickup_date"), "MMMM")
    )

    .withColumn(
        "is_weekend",
        F.when(
            F.dayofweek(F.col("tpep_pickup_datetime")).isin(1, 7),
            F.lit(1)
        ).otherwise(F.lit(0))
    )

    .withColumn(
        "day_type",
        F.when(F.col("is_weekend") == 1, F.lit("Weekend"))
         .otherwise(F.lit("Weekday"))
    )

    .withColumn(
        "time_period",
        F.when((F.col("pickup_hour") >= 5) & (F.col("pickup_hour") < 12), "Morning")
         .when((F.col("pickup_hour") >= 12) & (F.col("pickup_hour") < 17), "Afternoon")
         .when((F.col("pickup_hour") >= 17) & (F.col("pickup_hour") < 21), "Evening")
         .otherwise("Night")
    )

    .withColumn(
        "is_rush_hour",
        F.when(
            ((F.col("pickup_hour") >= 7) & (F.col("pickup_hour") <= 9)) |
            ((F.col("pickup_hour") >= 16) & (F.col("pickup_hour") <= 19)),
            F.lit(1)
        ).otherwise(F.lit(0))
    )

    # =========================
    # Distance-based features
    # =========================
    .withColumn(
        "distance_bucket",
        F.when(F.col("trip_distance") <= 1, "Very Short (0-1 mile)")
         .when(F.col("trip_distance") <= 2, "Short (1-2 miles)")
         .when(F.col("trip_distance") <= 5, "Medium (2-5 miles)")
         .when(F.col("trip_distance") <= 10, "Long (5-10 miles)")
         .when(F.col("trip_distance") <= 20, "Very Long (10-20 miles)")
         .otherwise("Extreme Valid (20+ miles)")
    )

    .withColumn(
        "fare_per_mile",
        F.round(F.col("fare_amount") / F.col("trip_distance"), 2)
    )

    .withColumn(
        "total_per_mile",
        F.round(F.col("total_amount") / F.col("trip_distance"), 2)
    )

    # =========================
    # Duration-based features
    # =========================
    .withColumn(
        "duration_bucket",
        F.when(F.col("trip_duration_minutes") <= 5, "Very Short (0-5 min)")
         .when(F.col("trip_duration_minutes") <= 15, "Short (5-15 min)")
         .when(F.col("trip_duration_minutes") <= 30, "Medium (15-30 min)")
         .when(F.col("trip_duration_minutes") <= 60, "Long (30-60 min)")
         .when(F.col("trip_duration_minutes") <= 120, "Very Long (60-120 min)")
         .otherwise("Extreme Valid (120+ min)")
    )

    # =========================
    # Speed feature
    # =========================
    .withColumn(
        "speed_category",
        F.when(F.col("average_speed_mph") < 8, "Very Slow")
         .when(F.col("average_speed_mph") < 20, "Normal City Speed")
         .when(F.col("average_speed_mph") < 40, "Fast City Trip")
         .when(F.col("average_speed_mph") < 60, "High Speed")
         .otherwise("Very High Speed")
    )

    # =========================
    # Tip behavior features
    # =========================
    .withColumn(
        "tip_base_amount",
        F.when(
            (F.col("payment_type") == 1) &
            ((F.col("total_amount") - F.col("tip_amount")) > 0),
            F.round(F.col("total_amount") - F.col("tip_amount"), 2)
        ).otherwise(None)
    )

    .withColumn(
        "tip_percentage",
        F.when(
            (F.col("payment_type") == 1) &
            (F.col("tip_base_amount") > 0),
            F.round((F.col("tip_amount") / F.col("tip_base_amount")) * 100, 2)
        ).otherwise(None)
    )

    .withColumn(
        "tip_category",
        F.when(F.col("payment_type") != 1, "Cash Tip Not Recorded")
         .when(F.col("tip_amount") == 0, "No Tip")
         .when(F.col("tip_percentage") < 10, "Low Tip")
         .when(F.col("tip_percentage") < 20, "Medium Tip")
         .otherwise("High Tip")
    )

    # =========================
    # Payment and rate labels
    # =========================
    .withColumn(
        "payment_type_label",
        F.when(F.col("payment_type") == 1, "Credit Card")
         .when(F.col("payment_type") == 2, "Cash")
         .otherwise("Other")
    )

    .withColumn(
        "ratecode_label",
        F.when(F.col("RatecodeID") == 1, "Standard Rate")
         .when(F.col("RatecodeID") == 2, "JFK")
         .when(F.col("RatecodeID") == 3, "Newark")
         .when(F.col("RatecodeID") == 4, "Nassau or Westchester")
         .when(F.col("RatecodeID") == 5, "Negotiated Fare")
         .when(F.col("RatecodeID") == 6, "Group Ride")
         .otherwise("Unknown")
    )

    # =========================
    # Airport trip feature
    # =========================
    .withColumn(
        "is_airport_trip",
        F.when(
            (F.col("RatecodeID").isin(2, 3)) | (F.col("airport_fee") > 0),
            F.lit(1)
        ).otherwise(F.lit(0))
    )

    # =========================
    # Route feature
    # =========================
    .withColumn(
        "route_pair",
        F.concat_ws(
            "_to_",
            F.col("PULocationID").cast("string"),
            F.col("DOLocationID").cast("string")
        )
    )

    # =========================
    # Financial consistency label
    # =========================
    .withColumn(
        "financial_consistency_label",
        F.when(F.col("is_total_amount_inconsistent") == True, "Inconsistent")
         .otherwise("Consistent")
    )
)
````

## Gold KPI Summary — General Purpose

The `gold_kpi_summary` table gives a high-level summary of the whole cleaned taxi dataset.

Instead of showing millions of trip records, this table provides the main KPIs after Silver cleaning.

It answers the question:

**What are the main KPIs of the cleaned taxi dataset?**

### Main KPIs Included

- Total clean trips
- Total recorded passenger charges
- Average fare and total amount
- Average trip distance, duration, and speed
- Credit card vs cash payment percentage
- Airport trip percentage
- Financial consistency percentage
- Average credit card tip percentage

### Key Insight

The `gold_kpi_summary` table is useful for the first dashboard page or report overview because it gives a quick business summary of the cleaned taxi data.

````python
%sql
CREATE OR REPLACE TABLE main.default.gold_kpi_summary AS
SELECT
    COUNT(*) AS total_trips,

    ROUND(SUM(total_amount), 2) AS total_recorded_passenger_charges,
    ROUND(SUM(fare_amount), 2) AS total_base_fare_amount,

    ROUND(AVG(total_amount), 2) AS avg_passenger_charge,
    ROUND(AVG(fare_amount), 2) AS avg_base_fare,

    ROUND(AVG(trip_distance), 2) AS avg_trip_distance,
    ROUND(AVG(trip_duration_minutes), 2) AS avg_duration_minutes,
    ROUND(AVG(average_speed_mph), 2) AS avg_speed_mph,

    ROUND(AVG(fare_per_mile), 2) AS avg_base_fare_per_mile,
    ROUND(AVG(total_per_mile), 2) AS avg_passenger_charge_per_mile,

    SUM(CASE WHEN payment_type = 1 THEN 1 ELSE 0 END) AS credit_card_trips,
    SUM(CASE WHEN payment_type = 2 THEN 1 ELSE 0 END) AS cash_trips,

    ROUND(
        SUM(CASE WHEN payment_type = 1 THEN 1 ELSE 0 END) * 100.0 / COUNT(*),
        2
    ) AS credit_card_trip_percentage,

    ROUND(
        SUM(CASE WHEN payment_type = 2 THEN 1 ELSE 0 END) * 100.0 / COUNT(*),
        2
    ) AS cash_trip_percentage,

    SUM(CASE WHEN is_airport_trip = 1 THEN 1 ELSE 0 END) AS airport_trips,
    SUM(CASE WHEN is_airport_trip = 0 THEN 1 ELSE 0 END) AS non_airport_trips,

    ROUND(
        SUM(CASE WHEN is_airport_trip = 1 THEN 1 ELSE 0 END) * 100.0 / COUNT(*),
        2
    ) AS airport_trip_percentage,

    ROUND(
        SUM(CASE WHEN is_airport_trip = 0 THEN 1 ELSE 0 END) * 100.0 / COUNT(*),
        2
    ) AS non_airport_trip_percentage,

    SUM(CASE WHEN financial_consistency_label = 'Consistent' THEN 1 ELSE 0 END) AS financially_consistent_trips,
    SUM(CASE WHEN financial_consistency_label = 'Inconsistent' THEN 1 ELSE 0 END) AS financially_inconsistent_trips,

    ROUND(
        SUM(CASE WHEN financial_consistency_label = 'Consistent' THEN 1 ELSE 0 END) * 100.0 / COUNT(*),
        2
    ) AS financially_consistent_percentage,

    ROUND(
        SUM(CASE WHEN financial_consistency_label = 'Inconsistent' THEN 1 ELSE 0 END) * 100.0 / COUNT(*),
        2
    ) AS financially_inconsistent_percentage,

    ROUND(AVG(CASE WHEN payment_type = 1 THEN tip_percentage END), 2) AS avg_credit_card_tip_percentage

FROM main.default.yellow_taxi_2023_features;
````

## Load Feature Table

````python
from pyspark.sql import functions as F

features_table = "main.default.yellow_taxi_2023_features"

df_features = spark.table(features_table)

print(f"Feature table rows: {df_features.count():,}")
display(df_features.limit(5))
````

## Select ML Columns

````python
ml_columns = [
    # Target
    "fare_amount",

    # Numeric / pre-trip features
    "trip_distance",
    "passenger_count",
    "pickup_month",
    "pickup_hour",
    "pickup_day_of_week",
    "is_weekend",
    "is_rush_hour",
    "is_airport_trip",

    # Categorical features
    "VendorID",
    "RatecodeID",
    "PULocationID",
    "DOLocationID",
    "time_period",
    "day_type",
    "distance_bucket"
]

df_ml = df_features.select(ml_columns).dropna()

print(f"ML dataset rows: {df_ml.count():,}")
display(df_ml.limit(10))
````

## Cast Categorical Columns to String

````python
df_ml_ready = (
    df_ml
    .withColumn("VendorID_str", F.col("VendorID").cast("string"))
    .withColumn("RatecodeID_str", F.col("RatecodeID").cast("string"))
    .withColumn("PULocationID_str", F.col("PULocationID").cast("string"))
    .withColumn("DOLocationID_str", F.col("DOLocationID").cast("string"))
)
````

## Train-Test Split

````python
train_df, test_df = df_ml_ready.randomSplit([0.8, 0.2], seed=42)

print(f"Training rows: {train_df.count():,}")
print(f"Testing rows: {test_df.count():,}")
````

## Build the ML Pipeline

````python
from pyspark.ml.feature import StringIndexer, OneHotEncoder, VectorAssembler
from pyspark.ml.regression import LinearRegression
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml import Pipeline
````

````python
numeric_features = [
    "trip_distance",
    "passenger_count",
    "pickup_month",
    "pickup_hour",
    "pickup_day_of_week",
    "is_weekend",
    "is_rush_hour",
    "is_airport_trip"
]

categorical_features = [
    "VendorID_str",
    "RatecodeID_str",
    "PULocationID_str",
    "DOLocationID_str",
    "time_period",
    "day_type",
    "distance_bucket"
]
````

````python
indexers = [
    StringIndexer(
        inputCol=col_name,
        outputCol=col_name + "_index",
        handleInvalid="keep"
    )
    for col_name in categorical_features
]
````

````python
encoders = [
    OneHotEncoder(
        inputCol=col_name + "_index",
        outputCol=col_name + "_encoded"
    )
    for col_name in categorical_features
]
````

````python
encoded_features = [col_name + "_encoded" for col_name in categorical_features]

assembler = VectorAssembler(
    inputCols=numeric_features + encoded_features,
    outputCol="features",
    handleInvalid="keep"
)
````

## Train Linear Regression Model

````python
lr = LinearRegression(
    featuresCol="features",
    labelCol="fare_amount",
    predictionCol="prediction"
)

lr_pipeline = Pipeline(
    stages=indexers + encoders + [assembler, lr]
)

lr_model = lr_pipeline.fit(train_df)

lr_predictions = lr_model.transform(test_df)

display(
    lr_predictions.select(
        "fare_amount",
        "prediction",
        "trip_distance",
        "pickup_hour",
        "pickup_day_of_week",
        "is_airport_trip",
        "PULocationID",
        "DOLocationID"
    ).limit(20)
)
````

## Evaluate Linear Regression

````python
rmse_evaluator = RegressionEvaluator(
    labelCol="fare_amount",
    predictionCol="prediction",
    metricName="rmse"
)

mae_evaluator = RegressionEvaluator(
    labelCol="fare_amount",
    predictionCol="prediction",
    metricName="mae"
)

r2_evaluator = RegressionEvaluator(
    labelCol="fare_amount",
    predictionCol="prediction",
    metricName="r2"
)

lr_rmse = rmse_evaluator.evaluate(lr_predictions)
lr_mae = mae_evaluator.evaluate(lr_predictions)
lr_r2 = r2_evaluator.evaluate(lr_predictions)

print("Linear Regression Results")
print(f"RMSE: {lr_rmse:.2f}")
print(f"MAE: {lr_mae:.2f}")
print(f"R2: {lr_r2:.4f}")
````

## Save Model Evaluation Results

````python
model_results = [
    ("Linear Regression", float(lr_rmse), float(lr_mae), float(lr_r2))
]

model_results_df = spark.createDataFrame(
    model_results,
    ["model_name", "rmse", "mae", "r2"]
)

display(model_results_df)

(
    model_results_df
    .write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable("main.default.gold_model_evaluation")
)

print("Model evaluation table saved successfully:")
print("main.default.gold_model_evaluation")
````
